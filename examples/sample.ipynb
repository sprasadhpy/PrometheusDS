```python
import os
import httpx
from getpass import getpass
import pandas as pd

from langchain_openai import ChatOpenAI
from Prometheus_DS.agents import DataCleaningAgent
```

```python
# ─── Prometheus API Configuration ─────────────────────
API_BASE = "https://api.entain-gpt-dev.aws.local"
API_KEY = os.environ.get("PROMETHEUS_API_KEY") or getpass("Enter your Prometheus API key: ")

# ─── GPT Model ─────────────────────
llm_gpt = ChatOpenAI(
    model="gpt-5.2",
    api_key="API_KEY",
    base_url=f"{API_BASE}/v1",
    http_client=httpx.Client(verify=False),
    default_headers={"x-api-key": API_KEY},
)

# ─── Qwen Model ─────────────────────
llm_qwen = ChatOpenAI(
    model="qwen-coder",
    api_key="API_KEY",
    base_url=f"{API_BASE}/v1",
    http_client=httpx.Client(verify=False),
    default_headers={"x-api-key": API_KEY},
)

# Choose which model to use
llm = llm_qwen  # Switch to llm_gpt if you want GPT

LOG = True
LOG_PATH = os.path.join(os.getcwd(), "logs/")

print(f"Active model: {llm.model_name}")
```
